```python
# SIMPLE LINEAR REGRESSIONS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

f= {'EXPERIENCE':[1.1,2.2,3.5,4.9,5.1,6.4],
    'SALARY':[10000,22000,38000,72000,89000,99000]}
df=pd.DataFrame(f)
df
```

```python
x=df['EXPERIENCE']
y=df['SALARY']
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=12)
x_train

```

```python
# TRAINING TO ALGORITHM
trained_data = pd.DataFrame({'X_trian':x_train,'Y_train':y_train})
trained_data
```

```python
# TESTING THE ALGORITHM
testing_data = pd.DataFrame({'X_test':x_test,'Y_test':y_test})
testing_data
```

```python
# GIVE THE TRAINING DATA
# BUT BEFORE X-VARIABLE CAN BE CONVERT INTO TWO DIMENSION
x_train = x_train.values.reshape(-1,1)
x_train.shape
```

```python
# HERE IMPORTING LINEAR REGRESSION CLASS
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
# what is m value
reg.coef_
```

```python
# what is c value
reg.intercept_
```

```python
trained_data
```

```python
#NOW FIND Y VALUE
# y=mx+c
# y = 17089.17197452 * x + -14144.267515923566)
latha_answers = reg.predict(x_train)
latha_answers
```
